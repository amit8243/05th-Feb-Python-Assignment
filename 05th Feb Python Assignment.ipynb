```python
#Python Assignment
"""Q1. Explain Class and Object with respect to Object-Oriented Programming. Give a suitable example."""
Ans: Class: A class is a blueprint or template for creating objects. It defines the data type of a particular object 
and the types of operations that can be performed on it.

Object: An object is an instance of a class. It is a real-world entity that has properties and behavior.

Example: A car is an example of a class. It has properties such as color, make, model, and year. It also has behaviors 
such as accelerating, braking, and turning.

An individual car, such as a red Ford Mustang from 2020, is an example of an object. It has the properties of color 
(red), make (Ford), model (Mustang), and year (2020). It also has the behaviors of accelerating, braking, and turning.

"""Q2. Name the four pillars of OOPs."""
Ans: 
1. Abstraction
2. Encapsulation
3. Inheritance
4. Polymorphism

"""Q3. Explain why the __init__() function is used. Give a suitable example."""
Ans: The __init__() function is used to initialize an object's state. It is called when an object of a class is created 
and can be used to set the attributes of the object. For example, if we have a class called 'Car', we can use 
the __init__() function to set the attributes of the car such as its make, model, color, etc.

class Car:
    def __init__(self, make, model, color):
        self.make = make
        self.model = model
        self.color = color

my_car = Car('Honda', 'Civic', 'Red')
print(my_car.make)
print(my_car.model)
print(my_car.color)

# Output: 
# Honda 
# Civic 
# Red
"""Q4. Why self is used in OOPs?"""
Ans: Self is used in OOPs to refer to the current instance of a class. It allows the programmer to access the attributes 
and methods of the class without having to explicitly specify an object. This makes it easier to write code that can be 
reused across multiple objects.

Self is also used to differentiate between class attributes and instance attributes. Class attributes are shared across
all instances of a class, while instance attributes are specific to each instance.

"""Q5. What is inheritance? Give an example for each type of inheritance."""
Ans: Inheritance is the process of passing down characteristics from one generation to the next. It is a fundamental 
concept in object-oriented programming, where classes can inherit attributes and behavior from other classes.

Examples of inheritance include: 
1. Single Inheritance: A class inheriting from a single parent class. For example, a Dog class inheriting from an Animal class. 
2. Multiple Inheritance: A class inheriting from multiple parent classes. For example, a FlyingFish class inheriting from both Fish and Bird classes. 
3. Multilevel Inheritance: A class inheriting from a parent class, which in turn inherits from a grandparent class. For example, a Labrador class inheriting from a Dog class, which in turn inherits from an Animal class. 
4. Hierarchical Inheritance: Multiple classes inheriting from the same parent class. For example, a Cat and Dog class both inheriting from an Animal class. 
5. Hybrid Inheritance: A combination of two or more types of inheritance. For example, a FlyingDog class inheriting from both Dog and Bird classes.
```
